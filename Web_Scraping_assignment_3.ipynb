2.In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“.

```python
Brand_Name=[]#
Product_Name=[] #
Ratings=[]
Total_Ratings=[] #
Price=[] #
Return_Exchange=[]#
Expected_Delivery=[]#
Availability=[]#
Product_Url=[]
```

```python
for pages in range(0,3):
    
    tags=driver.find_elements_by_xpath("//span[@class='a-size-base']")
    
    
    for i in tags:
        Total_Ratings.append(i.text)
        
    urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//li[@class='a-normal']/a")]      
    for url in urls:     
        driver.get(url)
        
    
driver.get('https://www.amazon.in/s?k=guitar&qid=1614188464&ref=sr_pg_1')       
print(Total_Ratings)
```

```python
for pages in range(0,3):
    
    tags=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
    
    
    for i in tags:
        Price.append(i.text)
        
    urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//li[@class='a-normal']/a")] 
    for url in urls:   
        driver.get(url)
        

driver.get('https://www.amazon.in/s?k=guitar&qid=1614188464&ref=sr_pg_1')   
print(Price)
```

```python
for pages in range(0,3):
    
    tags=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
    
    
    for i in tags:
        Product_Name.append(i.text)
   
    urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//li[@class='a-normal']/a")]    
    for url in urls:
        driver.get(url)
        
    
driver.get('https://www.amazon.in/s?k=guitar&qid=1614188464&ref=sr_pg_1')    
print(Product_Name)
```

```python
for pages in range(0,3):
    urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a")]    
    for url in urls:       
        driver.get(url)
        
        tags=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-top-mini']/span/b")
        for i in tags:
            Expected_Delivery.append(i.text)
        
        tags_2=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none  icon-content']/a")
        for i in tags_2:
            Return_Exchange.append(i.text)
     
        tags_3=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none']/a")
        for i in tags_3:
            Brand_Name.append(i.text)
        
        tags_4=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-success']")
        for i in tags_4:
            Availability.append(i.text)
        
    urls_2=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//li[@class='a-normal']/a")] 
    for url_2 in urls_2:   
        driver.get(url_2)        

            
driver.get('https://www.amazon.in/s?k=guitar&qid=1614188464&ref=sr_pg_1')         
print(Expected_Delivery)
```
