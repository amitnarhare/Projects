1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
# Activating the chrome browser
driver = webdriver.Chrome('D:\DataTrained\chromedriver_win32\chromedriver.exe')
```

```python
#path of wikipedia
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
#Rank
rk = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for d in rk:
    Rank.append(d.text)
#Name
nm = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for d in nm:
    Name.append(d.text)  
#Artist    
art = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for d in art:
    Artist.append(d.text)    
#Upload Date
dt = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for d in dt:
    Upload_date.append(d.text)  
#Views
vw = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for d in vw:
    Views.append(d.text)    
```

```python
TOP=pd.DataFrame({})
TOP['Rank']=Rank[:30]
TOP['Name']=Name[:30]
TOP['Artist']=Artist[:30]
TOP['Upload_date']=Upload_date[:30]
TOP['Views(Billion)']=Views[:30]
TOP
```

st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Activating the chrome browser
driver = webdriver.Chrome('D:\DataTrained\chromedriver_win32\chromedriver.exe')
```

```python
#BCCI Home Page path
driver.get(' https://www.bcci.tv/')
```

```python
lst=[]
urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']")]    
for i in urls:
    lst.append(i)
driver.get(lst[0])
```

```python
#Decleration of empty list
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]
lst=[]
```

```python
#Match Title
tt = driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]/p/strong')
for d in tt:
    Match_title.append(d.text)   
#Place    
pc = driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]/p/span')
for d in pc:
    Place.append(d.text)
#Date
ser = driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]/span[2]')
for d in ser:
    Series.append(d.text)
#Time
dt = driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]/span')
for d in dt:
    lst.append(d.text)   
```

```python
#Spliting the date and time
for i in range(0,len(lst)):
    x=lst[i].split()
    Time.append(x[3]+" "+x[4])
    Date.append(x[0]+" "+x[1]+" "+x[2])   
```

```python
Schedule=pd.DataFrame({})
Schedule['Match_title']=Match_title[:]
Schedule['Series']=Series[:]
Schedule['Place']=Place[:]
Schedule['Date']=Date[:]
Schedule['Time']=Time[:]
Schedule
```
