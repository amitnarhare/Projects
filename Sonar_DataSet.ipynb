```python
dum=pd.get_dummies(df['Class'],drop_first=True)
```

```python
dum
```

```python
df_new=pd.concat([df,dum],axis=1)

df_new=df_new.drop('Class',axis=1)
```

```python
df_new
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df_new))
```

```python
z
```

```python
print(np.where(z>3))
```

```python
df_new=df_new[(z<3).all(axis=1)]
df_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
x=df_new.drop('Rock',axis=1)
y=df_new['Rock']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
```

```python
print("accuracy:",accuracy_score(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```

```python
classification_report(y_test,pred)
```

```python
#Cross validation
from sklearn.model_selection import cross_val_score
dtscore=cross_val_score(LR,x,y,cv=5)
print('cross validation=',dtscore)
print(dtscore.mean(),dtscore.std())
```

```python
from sklearn.externals import joblib
joblib.dump(LR,'LR_Sonar.pkl')
```
